Question 1: Define Object Oriented Programming Language?

Answer: Object-oriented programming (OOP) refers to a type of computer programming (software design) in which programmers define the data type of a data structure, and also the types of operations (functions) that can be applied to the data structure.

Question 4: Define the following terms: 1. Class 2. Object 3. Attribute 4. Behavior.

Answer:

Class: The class is a blueprint that defines a nature of a future object

Object: an object can be a data structure, a variable or a function. It has a memory location allocated. The object is designed as class hierarchies.

Attribute: Attributes are data stored inside a class or instance and represent the state or quality of the class. One can think of attributes as noun or adjective, while methods are the verb of the class.

Behavior: Behaviors are the methods corresponding to the behavior of real world objects

```python
#Question 4: Write a code in python in which create a class named it Car which have 5 attributes such like (model, color and name etc.) and 3 methods. And create 5 object instance from that class.

class Car():

    def __init__(self, name, model, color, company, transmission):
        self.name = name
        self.model = model
        self.color = color
        self.company = company
        self.transmission = transmission
        

    def show_properties(self):
        print("Car Properties")
        print(f"Name         : {self.name}")
        print(f"Model        : {self.model}")
        print(f"Color        : {self.color}")
        print(f"Company      : {self.company}")
        print(f"Transmission : {self.transmission} \n")
        
        
    def update_name(self, newname):
        self.name = newname
    

    def update_color(self, newcolor):
        self.color = newcolor


car1 = Car("BRV", 2019, "Red", "Honda", "Auto")
car2 = Car("Alto", 2008, "grey", "Suzuki", "Manual")
car3 = Car("Corolla", 2016, "Black", "Toyota", "Auto")
car4 = Car("Civic", 2007, "Grey", "Honda", "Manual")
car5 = Car("City", 2015, "White", "Honda", "Auto")


car1.show_properties()
car2.show_properties()
car3.show_properties()
car4.show_properties()
car5.show_properties()
        
    
print("Before Changing : ",car1.name)
car1.update_name("Land Cruiser")
print("After Changing : ",car1.name)
```
